## Q1. Total number of people vaccinated in Argentina

```python
df["country"].unique()
```

```python
ad=df[df["country"]=="Argentina"]
ad
```

```python
# considering total_vaccinations as a column where all people who are partially vaccinated and  fully vaccinated  are summed up together

ad.groupby("country").sum()["total_vaccinations"]
```
